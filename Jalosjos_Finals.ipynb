```python
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
from sklearn.metrics import r2_score
import ipywidgets as widgets
from IPython.display import display
from scipy.stats import skew, zscore
```

```python
# Read the data set from data folder
data = pd.read_csv("data/medical_insurance_dataset.csv")
```

```python
# No region column 
columns_used = ['age', 'bmi', 'children', 'charges', 'sex', 'smoker']
number_variables = ['age', 'bmi', 'children', 'charges']
# statuses = {'sex', 'smoker'}
specific_choice = ['age', 'smoker']
gender_list = ['male', 'female']
status_of_smoker = ['yes', 'no']

central_tendencies = data.loc[:, data.columns.isin(number_variables)]
general_data = data.loc[:, data.columns.isin(columns_used)]
# sex_and_smoker_status = data.loc[:, data.columns.isin(statuses)]
```

```python
def measure_of_central_tendencies(central_tendencies, feature):

    # Measures of Central Tendencies
    mean = central_tendencies[feature].mean()
    median = central_tendencies[feature].median()
    mode = central_tendencies[feature].mode()[0]

    print("\nCentral Tendency Measures")
    print(f"Mean: {mean: .4f}")
    print(f"Median: {median: .4f}")
    print(f"Mode: {mode: .4f}")
    
    # Measures of Dispersion
    variance = central_tendencies[feature].var()
    stddev = central_tendencies[feature].std()
    range = central_tendencies[feature].max() - central_tendencies[feature].min()
    q1 = central_tendencies[feature].quantile(0.25)
    q3 = central_tendencies[feature].quantile(0.75)
    iqr = q3 - q1

    print("\nMeasures of Dispersion")
    print(f"Variance: {variance: .4f}")
    print(f"Standard Deviation: {stddev: .4f}")
    print(f"Range: {range: .4f}")
    print(f"Q1 of Data: {q1: .4f}")
    print(f"Q3 of Data: {q3: .4f}")
    print(f"Interquartile Range: {iqr: .4f}")
    
    # Skewness
    skewness = skew(central_tendencies[feature])

    print("\nSkewness")
    print(f"Skewness: {skewness: .4f}")

    # Outliers
    zscores = np.abs(zscore(central_tendencies[feature]))
    outlier_indices = np.where(zscores > 2.5)[0]

    print("\nOutliers")
    print(central_tendencies[feature][outlier_indices])
```

```python
def data_display(category):
    print(f"Measure of Central Tendencies of: {category}")
    measure_of_central_tendencies(central_tendencies, category)
```

```python
widgets.interact(data_display, category = number_variables)
```

```python
# Create a sub-data set for the smokers and sex 

m_smokers = general_data[(general_data["sex"] == "male") & (general_data["smoker"] == "yes")]
m_non_smokers = general_data[(general_data["sex"] == "male") & (general_data["smoker"] == "no")]

f_smokers = general_data[(general_data["sex"] == "female") & (general_data["smoker"] == "yes")]
f_non_smokers = general_data[(general_data["sex"] == "female") & (general_data["smoker"] == "no")] 
```

```python
# Remove the string values + finalize the subset

dropped_m_smokers = m_smokers.drop(['sex', 'smoker'], axis=1) 
dropped_m_non_smokers = m_non_smokers.drop(['sex', 'smoker'], axis=1)
dropped_f_smokers = f_smokers.drop(['sex','smoker'], axis=1) 
dropped_f_non_smokers = f_non_smokers.drop(['sex', 'smoker'], axis=1) 

dropped_m_smokers.name = "Male Smokers" 
dropped_m_non_smokers.name = "Male Non-Smokers" 
dropped_f_smokers.name = "Female Smokers" 
dropped_f_non_smokers.name = "Female Non-Smokers" 
```

```python
def linear_regression(x_input, y_input, data_input): 
    X = data_input[[x_input]]
    Y = data_input[[y_input]]

    model = LinearRegression().fit(X, Y)
    predictions = model.predict(X)

    intercept = list(map('{:.2f}' .format, model.intercept_))
    coef = list(map('{:.2f}' .format, model.coef_[0]))

    plt.figure(figsize=(8, 8))
    sns.scatterplot(data=data_input, x=x_input, y=y_input)
    line = model.intercept_ + model.coef_[0] * data_input[x_input]
    equation = f'Charge = {intercept} + {coef} * {x_input}'
    plt.plot(data_input[[x_input]], line, 'r', label=f"y={intercept}+{coef}*x")

    print(f"\n The Regression Equation for {x_input}: ")
    print(equation)

    # Scatterplot
    plt.legend()
    plt.title(f"The Regression of {x_input} on Charge")
    plt.grid(True)
    plt.show()
```

```python
def multilinear_calculation(x_input, y_input, z_input, a_input, data_input): 
    X = data_input[[y_input, z_input, a_input]]
    Y = data_input[x_input]

    model = LinearRegression().fit(X, Y)
    predictions = model.predict(X)

    r2score_corr = r2_score(Y, predictions)

    intercept = model.intercept_
    coef = model.coef_

    equation = f'Charge = {intercept} + {coef[0]} * Age + {coef[1]} * BMI + {coef[2]} * Children'

    print(f"\n The Rregression Equation for {x_input}: ")
    print(equation) 
    print(f"R-Squared: {r2score_corr}")

    sns.pairplot(data_input, kind="reg", plot_kws={'line_kws':{'color':'red'}})

    return [coef, intercept]
```

```python
# Create a list of all final data subsets
variables = [dropped_m_smokers, dropped_m_non_smokers, dropped_f_smokers, dropped_f_non_smokers]

for var in variables:
    linear_regression('age', 'charges', var)
    linear_regression('bmi', 'charges', var)
    multilinear_calculation('charges', 'age', 'bmi', 'children', var)
```

```python
def final_part(sex, smoker, age_input, bmi_input, children_input): 
    data_input = general_data[general_data["sex"] == sex]
    data_input = data_input[data_input["smoker"] == smoker]

    regression = multilinear_calculation('charges', 'age', 'bmi', 'children', data_input)
    intercept = regression[1]
    coef = regression[0]
    
    age_input = float(age_input.replace(',',''))
    bmi_input = float(bmi_input.replace(',',''))
    children_input = float(children_input.replace(',',''))
    input_prediction = intercept + coef[0] * age_input + coef[1] * bmi_input + coef[2] * children_input
    print(f"Charged Prediction from inputs: {input_prediction:.2f}")
```

```python
widgets.interact(final_part, sex = gender_list, smoker = status_of_smoker, age_input = '0', bmi_input = '0', children_input = '0')
```
